```python
# Build regression model
model = smf.ols('sunday~daily',data = data1).fit()
model.summary()
```

#### Observations
- The predicted equation y_hat = beta_0 + beta_1*x 
- y_hat = 13.356 + 1.3397x

- The probability(p-value) for intercept (beta_0) is 0.707 > 0.05
- Therefore the intercept coefficient may not be that much signficant in prediction
- However the p-value for 'daily' (beta_1) is 0.00 < 0.05
- Therfore the beta_1 coefficient is highly significant and is contributint to prediction.

```python
x = data1['daily'].values
y = data1['sunday'].values
plt.scatter(x, y, color = 'm', marker = 'o', s = 30)
b0 = 13.84
b1 = 1.33
y_hat = b0 + b1*x
plt.plot(x, y_hat, color = 'g')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
model1 = smf.ols("sunday~daily",data= data1).fit()
model1.summary()
```

```python
model1.params
```

```python
print(f'model t-values:\n{model1.tvalues}\n-----------------\nmodel p-value: \n{model1.pvalues}')
```

#### predict for new data

```python
newdata=pd.Series([200,300,400])
data_pred = pd.DataFrame(newdata, columns=['daily'])
data_pred
```

```python
model1.predict(data_pred)
```

```python
pred = model1.predict(data1['daily'])
pred
```

```python
data1['Y_hat'] = pred
data1
```

```python
data1['Y_hat'] = pred
data1
```

```python
mse = np.mean((data1["daily"]-data1["Y_hat"])**2)
rmse = np.sqrt(mse)
print("MSE: ",mse)
print("RMSE: ",rmse)
```
